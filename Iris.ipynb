### Задачи:

Сколько видов ирисов в датасете, какие это виды?

Найти средние и медианы размеры чашелистиков и лепестков для каждого из видов (построить сводную таблицу).


```python
import pandas as pd
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv') # открытие датасета
iris.head()
```

```python
# Сколько видов ирисов в датасете, какие это виды?
iris.species.value_counts()
```

##### В датасете есть три вида ирисов: setosa, versicolor, virginica.

```python
# Средние и медианы размеры чашелистиков и лепестков для каждого из видов.
iris.groupby('species').agg({
    'sepal_length':['mean','median'],# среднее и медиана длины чашелистиков
    'sepal_width':['mean','median'],# среднее и медиана ширины чашелистиков
    'petal_length':['mean','median'],# среднее и медиана длины лепестков
    'petal_width':['mean','median']# среднее и медиана ширины лепестков
})
```
